```python
def pockemans(num):
    pockeman_map={
    1:"Q",
    2:"W",
    3:"E",
   
    }
    for i in num:
      if pockeman_map.get(i) is None:
         print("pockeman not found")
      else:
         print(pockeman_map[i])
```

```python
num=[int(x) for x in input().split()]
pockemans(num) 
```

```python

import matplotlib.pyplot as plt
    
```

```python
x=np.array([1,2,3,4,5])
y=x**2+3
print(x,y)
```

```python
plt.style.use('dark_background')
```

```python
plt.plot(x,x**2+3,label="Stock Price")
plt.plot(x,x**2,label="Stock Price")
plt.title("My Line Plot")
plt.xlabel("Time Stamp")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.scatter(x,x**2+3,label="Stock Price")
plt.scatter(x,x**2,label="Stock Price")
plt.title("My scatter Plot")
plt.xlabel("Time Stamp")
plt.ylabel("Price")
plt.legend()
plt.show()
```
